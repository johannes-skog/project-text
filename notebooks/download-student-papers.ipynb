```python
import time 

def lup_extract():
    
    limit_page = 100
    
    start = 0
    
    while 1:
        
        print(start)
        
        url = f"https://lup.lub.lu.se/student-papers/search/publication?limit={limit_page}&q=language+exact+swe&q=fulltext+exact+1&start={start}"
        
        start +=limit_page
        
        page = urllib.request.urlopen(url)
        
        page_html = page.read().decode("utf-8")
        
        # Find all the publication ids on the list-page
        publications = re.findall('https:\/\/lup.lub.lu.se\/student-papers\/search\/publication\/\d+', page_html)
        publications_ids = [x.split("/")[-1] for x in publications]
        
        if len(publications_ids) == 0:
            return 
        
        for publications_id in publications_ids:
            
            time.sleep(0.2)
            
            publication_url = f"https://lup.lub.lu.se/student-papers/search/publication/{publications_id}"
            
            try:
                # Go into the publication page for the publications_id
                
                publication_page = urllib.request.urlopen(
                   publication_url
                )
            except Exception as e:
                print("failed publication page:\n", publication_url)
                continue
            
            # Find all the pdf:s for this publication 
            pdf_paths = re.findall(
                f"http:\/\/lup.lub.lu.se\/student-papers\/record\/{publications_id}\/file\/\d+.pdf",
                publication_page.read().decode("utf-8")
            )
            
            publication_page.close()
            
            pdf_paths = np.unique(pdf_paths)
            
            # Download all the pdf:s 
            for pdf_path in pdf_paths:
                
                pdf_id = (pdf_path.split("/")[-1]).replace(".pdf", "")

                time.sleep(0.2)

                try:
                    urllib.request.urlretrieve(pdf_path, f"download_data/lup/{publications_id}_{pdf_id}.pdf")
                except Exception as e:
                    print("failed to download:\n", pdf_path)
                
        page.close()
```

```python
page_html = lup_extract()
```
